# Data gathering
From liveauctioneer.com

When searching for 'contemporary painting', the number of pages of search results appears to be > 300. However, results are only displayed for the first 10 pages and errors are shown for the remaining pages.

Prices are dynamically injected with javascript, so simply getting the html with requests does not seem to work. We could use more advanced web scrapers like selenium, but since there are only 10 pages of data, I manually copied those html files.

```python
import requests, re, time
# For HTML parsing
from bs4 import BeautifulSoup as BSHTML
# For making csv files to store prices of paintings
import pandas as pd
```

```python
# Change page number ('page=')
# pageSize only works after clicking manually, so use default of 24
example = 'https://www.liveauctioneers.com/search/?keyword=contemporary%20painting&page=2&pageSize=24&sort=-hammerPrice&status=archive'
# Turns out the prices are dynamically loaded so getting the html using requests does not work
```

```python
img = []
p = []
count = 0
for i in range(1, 10):
  # Manually copied sources
  example = open(f'/content/drive/My Drive/CPEN 291/project/liveauctioneer/p{i}.html')
  soup = BSHTML(example)
  images = soup.findAll('img', class_=['ImageWithHover__PrimaryImageWithFallback-sc-1b860cz-1', 'ImageRow__StyledImageWithFallback-sc-1s4yel2-3'])
  prices = soup.select("span[class^=PriceRow] > span[class^=FormattedCurrency__]")

  for price in prices:
    p.append(price.text)
    
  for image in images:
    img.append(image['src'])
    if ('https' in image['src']):
      count += 1


print(len(img))
print(len(p))
print(count)

img_final = []
p_final = []
for i in range(0, 1000):
  if ('https' in img[i]):
    img_final.append(img[i])
    p_final.append(p[i])
  
  
```

```python
# Create the pandas DataFrame
df = pd.DataFrame()
df.insert(loc=0, column='image', value=img_final)
df.insert(loc=0, column='price', value=p_final)
```
